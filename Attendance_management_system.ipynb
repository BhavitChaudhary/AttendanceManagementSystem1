```python
import sys
u = ('Bhavit', 1234)

print('Press 1 to login')
print('Press 2 to register')
ch = int(input("Your choice: "))

if ch == 1:
    x = input("Enter your name: ")
    y = int(input("Enter your password: "))
    if x == u[0] and y == u[1]:
        print("Welcome")
    else:
        print("Not authenticated. Exiting the program.")
        sys.exit()

def save_data_to_file(file_name, data):
    with open(file_name, 'w') as f:
        for name, info in data.items():
            roll_number, dob, address = info
            f.write(f"{name},{roll_number},{dob},{address}\n")

def load_data_from_file(file_name):
    data = {}
    try:
        with open(file_name, 'r') as f:
            for line in f:
                name, roll_number, dob, address = line.strip().split(',')
                data[name] = (roll_number, dob, address)
    except FileNotFoundError:
        pass
    return data

def create_student_database():
    student_database = load_data_from_file('students_database.txt')

    if not student_database:
        for i in range(5):
            name = input(f"Enter the name of student {i + 1}: ")
            roll_number = input(f"Enter the roll number of student {i + 1}: ")
            if roll_number in [data[0] for data in student_database.values()]:
                    print("Roll number already exists. Please try again.")
            else:
                 break
            dob = input(f"Enter the date of birth of student {i + 1}: ")
            address = input(f"Enter the address of student {i + 1}: ")

            student_database[name] = (roll_number, dob, address)

        save_data_to_file('students_database.txt', student_database)

    return student_database

def mark_attendance(students):
    attendance_records = load_data_from_file('attendance_records.txt')

    print("\nMark attendance for each student:")
    for name in students.keys():
        status = input(f"{name}: [P]resent / [A]bsent: ")
        if status.upper() == 'P':
            attendance_records[name] = 'Present'
        elif status.upper() == 'A':
            attendance_records[name] = 'Absent'
        else:
            print("Invalid input. Assuming 'Absent'.")
            attendance_records[name] = 'Absent'

    save_data_to_file('attendance_records.txt', attendance_records)

    return attendance_records

def view_attendance(students):
    attendance_records = load_data_from_file('attendance_records.txt')

    print("\nAttendance Records:")
    for name, info in students.items():
        roll_number, dob, address = info
        attendance_status = attendance_records.get(name, 'Absent')
        print(f"Name: {name}, Roll Number: {roll_number}, Date of Birth: {dob}, Address: {address}, Attendance: {attendance_status}")

# Create the database of students
students_attendance = create_student_database()

while True:
    print("\nAttendance Management System")
    print("Press 1 to add a student")
    print("Press 2 to mark attendance")
    print("Press 3 to view attendance")
    print("Press 4 to generate attendance report")
    print("Press 5 to exit")

    choice = int(input("Your choice: "))


    if choice == 1:
        name = input("Enter the name of the student: ")
        roll_number = input("Enter the roll number of the student: ")
        if roll_number in [data[0] for data in students_attendance.values()]:
                print("Roll number already exists. Please enter a unique roll number.")
        else:
             break
        dob = input("Enter the date of birth of the student: ")
        address = input("Enter the address of the student: ")

        students_attendance[name] = (roll_number, dob, address)
        save_data_to_file('students_database.txt', students_attendance)
        print("Student added successfully!")

    elif choice == 2:
        mark_attendance(students_attendance)
        print("Attendance marked successfully!")

    elif choice == 3:
        view_attendance(students_attendance)

    elif choice == 4:
        # Generate attendance report (can be implemented as per requirement)
        print("Attendance report generated successfully!")

    elif choice == 5:
        print("Exiting the Attendance Management System.")
        break

    else:
        print("Invalid choice. Please try again.")
```
